```python
import json

try:
    ## JSON file 읽기
    rfile = open('data/usagov_bitly.txt','r',encoding='utf-8')
    lines = rfile.readlines()       # lines 리스트
    # print(type(lines),len(lines))
    
    ## JSON 디코딩 : json.load()
    ## lines(json문자열) -> python dict 객체
    rows = []
    for row in lines:
        rows.append(json.loads(row))
    # rows = [json.loads(row) for row in lines]
    print('rows : ',len(rows))
    print('rows[0]의 타입 : ',type(rows[0]))
    for row in rows[:10]:
        print(row['hh'])
except Exception as e:
    print('Error 발생 : ',e)
finally:
    rfile.close()
    
## dict -> DataFrame 변환
import pandas as pd
record_df = pd.DataFrame(rows)
print(record_df.info())
```

```python
### 공공데이터 사이트
# data.go.kr

### 교통사고 파일을 열어서 DataFrame으로 생성
## 1. 생성된 DataFrame을 통해서 총 사건수, 사망자수, 중상자수, 경상자수 조회
## 2. 서울 지역 사망자수를 조회하여 총 사망자수를 조회
import pandas as pd
import numpy as np

acc = pd.read_csv('data/교통사고 20201231.csv')

print('총 사건수 : ',np.sum(acc['사고건수']))
print('총 사망자수 : ',np.sum(acc['사망자수']))
print('총 중상자수 : ',np.sum(acc['중상자수']))
print('총 경상자수 : ',np.sum(acc['경상자수']))

s_tot = 0
for i in range(len(acc)-1):
    if acc.시도[i] == '서울':
        s_tot += acc.사망자수[i]
print('서울 총사망자수 : ',s_tot)
```

```python
jumsu = {
'홍길동':[50,60,50],
'이순신':[60,60,60],
'강감찬':[100,90,90],
'유관순':[80,90,90],
'강환석':[95,95,95]
}

## 4개의 Series로 만들어주세요
## 이름(name_s), 각 리스트 0번은 국어(kor_s), 리스트 1번은 영어(eng_s), 리스트 2번은 수학(math_s)
# 이름 시리즈
na = [key for key in jumsu]
name_s = pd.Series(na)
# 국어 시리즈
kor = [jumsu[key][0] for key in na]
kor_s = pd.Series(kor)
# 영어 시리즈
eng_s = pd.Series([jumsu[key][1] for key in na])
# 수학 시리즈
math_s = pd.Series([jumsu[key][2] for key in na])
# print(name_s)
# print(kor_s)
# print(eng_s)
# print(math_s)

## DataFrame 만들자
score = pd.DataFrame({'이름':name_s,'국어':kor_s,'영어':eng_s,'수학':math_s})

# 국어, 영어, 수학 합계 시리즈를 만드세요
total = kor_s + eng_s + math_s
print('===합계===')
print(total)

# 평균값 구하기
avg = round(total/3,2)
print('===평균===')
print(avg)

score['합계'] = total
score['평균'] = avg
score
```
